## Transitioning From Ephemeral Apps to Deployed Apps

```python
# You can also run "modal deploy -m pricer_service" at the command line in an activated environment
# pricer_service is the name of the python file
!modal deploy -m pricer_service
```

```python
# once deployed, the function can be created simply like below :
pricer = modal.Function.from_name("product-pricer-llama-service", "price")
```

```python
# This can take a while! We'll use faster approaches shortly

pricer.remote("Quadcast HyperX condenser mic, connects via usb-c to your computer for crystal clear audio")
```

```python
# Below is the better version of the service, where model is loaded into memory only once and then reused from cache

!modal deploy -m pricer_service2
```

```python
# When using the class based approach, which pre-loads model using @app.cls , below code is used to call the service
# Pricer is the class name
Pricer = modal.Cls.from_name("product-pricer-llama-service", "Pricer")
pricer = Pricer()
reply = pricer.price.remote("Quadcast HyperX condenser mic, connects via usb-c to your computer for crystal clear audio")
print(reply)
```
